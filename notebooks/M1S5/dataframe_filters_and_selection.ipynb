Aplique filtros para extrair informações específicas dos DataFrames.

```python
import pandas as pd
import numpy as np
```

```python
clientes_df = pd.read_csv('../../data/input/csv/clientes.csv')
pedidos_df = pd.read_csv('../../data/input/csv/pedidos.csv')
produtos_df = pd.read_csv('../../data/input/csv/produtos.csv')
```

```python
clientes_df_30_mais = clientes_df[clientes_df['idade'] > 30]
clientes_df_30_mais
```

Dentro dos dados da tabela de clientes, temos 6 clientes maiores que 30 anos.

```python
pedidos_df_maiores = pedidos_df[pedidos_df['valor_total'] > 500]
pedidos_df_maiores
```

Nos registros, buscamos aqueles pedidos que são maiores que R\$ 500,00 reais e obtivemos 4 entradas.

```python
eletronicos = produtos_df[produtos_df['categoria'].str.contains('Eletrônicos')]
eletronicos
```

Filtramos os produtos da categoria "Eletrônicos" e tivemos os produtos listados acima.

```python
clientes_df_30_mais.loc[:,['cidade']]
```

As cidades onde estão cituados os clientes com mais de 30 anos estão listados acima.

```python
eletronicos = eletronicos.sort_values('preco')
eletronicos.iloc[0:3, 1:4]
```

Os 3 eletrônicos mais baratos da tabela de produtos, estão listados acima.
